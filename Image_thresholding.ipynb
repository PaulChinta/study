# Thresholding

Grayscale Images: A grayscale (or graylevel) image is simply one in which the only colors are shades of gray. Often,the grayscale intensity is stored as an 8-bit integer giving 256 possible different shades of gray from black to white. 

Binary Image: Binary images are images whose pixels have only two possible intensity values. They are normally displayed as black and white. Numerically, the two values are often 0 for black, and either 1 or 255 for white.

In digital image processing, thresholding is the simplest method of segmenting images. From a grayscale image, thresholding can be used to create binary images.

## Load Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Binary Thresholding

```python
# import opencv
import cv2
 
# Read image
src = cv2.imread("Images/input_image.jpeg", cv2.IMREAD_GRAYSCALE)

# Set threshold and maxValue
thresh = 20
maxValue = 255
 
# Basic threshold example
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY);

print("original image crop \n", src[50:55, 300:310])
print("After Binary threshold crop \n", dst[50:55, 300:310])

plt.figure(figsize=[10,10])
plt.subplot(1,2,1);
plt.imshow(src, cmap = "gray");
plt.title("Original Image");
plt.subplot(1,2,2);
plt.imshow(dst, cmap = "gray");
plt.title("Thresholded Image");
```

## Inverse Binary Thresholding

Inverse-Binary Thresholding is just the opposite of Binary Thresholding. 

The destination pixel is set to:
    
    zero, if the corresponding source pixel is greater than the threshold
    maxValue, if the source pixel is less than the threshold
    
    

Thresholding Rule:

    if src(x,y) > thresh
      dst(x,y) = 0
    else
      dst(x,y) = maxValue

```python
# import opencv
import cv2
 
# Read image
src = cv2.imread("Images/input_image.jpeg", cv2.IMREAD_GRAYSCALE)

# Set threshold and maxValue
thresh = 70
maxValue = 255
 
# Basic threshold example
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY_INV);

print("original image crop \n", src[290:305, 170:185])
print("After Binary threshold crop \n", dst[290:305, 170:185])

plt.figure(figsize=[10,10])
plt.subplot(1,2,1);
plt.imshow(src, cmap = "gray");
plt.title("Original Image");
plt.subplot(1,2,2);
plt.imshow(dst, cmap = "gray");
plt.title("Thresholded Image");
```

## Truncate Thresholding

In this type of thresholding, the destination pixel is set to:
    
    threshold (thresh), if the corresponding source pixel is greater than the threshold
    otherwise, the source pixel value remains same.
    maxValue is ignored

Thresholding Rule:

    if src(x,y) > thresh
      dst(x,y) = thresh
    else
      dst(x,y) = src(x,y)

```python
# import opencv
import cv2
 
# Read image
src = cv2.imread("Images/input_image.jpeg", cv2.IMREAD_GRAYSCALE)

# Set threshold and maxValue
thresh = 127
 
# Basic threshold example
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TRUNC);

print("original image crop \n", src[50:55, 300:310])
print("After Binary threshold crop \n", dst[50:55, 300:310])

plt.figure(figsize=[10,10])
plt.subplot(1,2,1);
plt.imshow(src, cmap = "gray");
plt.title("Original Image");
plt.subplot(1,2,2);
plt.imshow(dst, cmap = "gray");
plt.title("Thresholded Image");
```

## Threshold to Zero

In this type of thresholding, the destination pixel is set to:

    source pixel value, if the corresponding source pixel is greater than the threshold
    0, if the corresponding source pixel value is less than threshold
    maxValue is ignored

Thresholding Rule:

    if src(x,y) > thresh
       dst(x,y) = src(x,y)
    else
       dst(x,y) = 0

```python
# import opencv
import cv2
 
# Read image
src = cv2.imread("Images/input_image.jpeg", cv2.IMREAD_GRAYSCALE)

# Set threshold and maxValue
thresh = 127
 
# Basic threshold example
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TOZERO);

# print("original image crop \n", src[50:55, 300:310])
# print("After Binary threshold crop \n", dst[50:55, 300:310])

plt.figure(figsize=[10,10])
plt.subplot(1,2,1);
plt.imshow(src, cmap = "gray");
plt.title("Original Image");
plt.subplot(1,2,2);
plt.imshow(dst, cmap = "gray");
plt.title("Thresholded Image");
```
